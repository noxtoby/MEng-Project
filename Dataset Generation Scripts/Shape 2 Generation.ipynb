```python
import svgwrite
import random
from svglib.svglib import svg2rlg
from reportlab.graphics import renderPM
import math
TRIHEIGHT = math.sqrt(3)/2.0
```

```python
def draw_triangle(insert, size, fill, rotate=None):
    x, y = insert
    points = [insert, (x + size, y), ((x + size / 2.0), (y + size * TRIHEIGHT))]
    triangle = dwg.add(dwg.polygon(points, fill=fill, stroke=svgwrite.rgb(0, 0, 0, '%')))
    if rotate:
            triangle.rotate(rotate, center=insert)
```

```python
def shape_create():
    x_min = random.randrange(20,110)
    y_min = random.randrange(20,110)
    choice = random.randrange(0,3)

    
    if choice == 0:
        return 0, x_min, y_min
    if choice == 1:
        return 1, x_min, y_min
    if choice == 2:
        return 2, x_min, y_min
    
    
```

```python
name = "shape.svg"
svg_size_width = 150
svg_size_height = 150
count_0 = 0
count_1 = 0
count_2 = 1000

rec_size = 30
rad_circ = 15

while(count_0 != 0 or count_1 != 0 or count_2 != 0):
    
    pos, x_min, y_min = shape_create()
    if pos == 0:
        if count_0 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.rect((x_min, y_min), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='black')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name = "fillsquare" + new_name + str(count_0) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_0 = count_0 -1
        
    if pos == 1:
        if count_1 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        draw_triangle((x_min,y_min), 30, 'black', rotate = 60)
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name = "filltriangle" + new_name + str(count_1) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_1 = count_1 -1
    if pos == 2:
        if count_2 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.circle(center=(x_min,y_min),
            r=rad_circ, 
            stroke=svgwrite.rgb(0, 0, 0, '%'),
            fill='white')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name = "circle" + new_name + str(count_2) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_2 = count_2 -1
    
```
