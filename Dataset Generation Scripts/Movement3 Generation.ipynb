```python
def shape_create():
    x_min = random.randrange(10,90)
    y_min = random.randrange(10,90)
    x_min2 = random.randrange(10,90)
    y_min2 = random.randrange(10,90)
    x_min3 = random.randrange(10,90)
    y_min3 = random.randrange(10,90)
    rec_size = 20
    x_max = x_min + rec_size
    y_max = y_min + rec_size
    x_max2 = x_min2 + rec_size
    y_max2 = y_min2 + rec_size
    x_max3 = x_min3 + rec_size
    y_max3 = y_min3 + rec_size

    #x1 not inside x2
    if x_min <= x_max2:
        if x_min >= x_min2:
            if y_min <= y_max2:
                if y_min >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_min <= x_max2:
        if x_min >= x_min2:
            if y_max <= y_max2:
                if y_max >= y_min2:
                    return 0, x_min, y_min,x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max <= x_max2:
        if x_max >= x_min2:
            if y_min <= y_max2:
                if y_min >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max <= x_max2:
        if x_max >= x_min2:
            if y_max <= y_max2:
                if y_max >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    
    #x1 not inside x3
    if x_min <= x_max3:
        if x_min >= x_min3:
            if y_min <= y_max3:
                if y_min >= y_min3:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_min <= x_max3:
        if x_min >= x_min3:
            if y_max <= y_max3:
                if y_max >= y_min3:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max <= x_max3:
        if x_max >= x_min3:
            if y_min <= y_max3:
                if y_min >= y_min3:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max <= x_max3:
        if x_max >= x_min3:
            if y_max <= y_max3:
                if y_max >= y_min3:
                    return 0, x_min, y_min,x_min2, y_min2, x_min3, y_min3, rec_size
    
    #x2 not inside x3
    if x_min3 <= x_max2:
        if x_min3 >= x_min2:
            if y_min3 <= y_max2:
                if y_min3 >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_min3 <= x_max2:
        if x_min3 >= x_min2:
            if y_max3 <= y_max2:
                if y_max3 >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max3 <= x_max2:
        if x_max3 >= x_min2:
            if y_min3 <= y_max2:
                if y_min3 >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if x_max3 <= x_max2:
        if x_max3 >= x_min2:
            if y_max3 <= y_max2:
                if y_max3 >= y_min2:
                    return 0, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    #sadasdf
    if y_min - y_max2 <= 6:
        if y_min - y_max2 >= 2:
            if abs(x_min - x_min2) <= 2:
                if y_min3 - y_max <= 6:
                    if y_min3 - y_max >= 2:
                        if abs(x_min3 - x_min) <= 2:
                            return 3, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if y_min2 - y_max <= 6:
        if y_min2 - y_max >= 2:
            if abs(x_min - x_min2) <= 2:
                return 2, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    if y_min - y_max2 <= 6:
        if y_min - y_max2 >= 2:
            if abs(x_min - x_min2) <= 2:
                return 2, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size  


    
    return 1, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size
    
    
```

```python
name = "movement.svg"
svg_size_width = 120
svg_size_height = 120
count_1 = 1000
count_2 =1000
count_3 =1000
while(count_1 != 0 or count_2 != 0 or count_3 != 0):

    pos, x_min, y_min, x_min2, y_min2, x_min3, y_min3, rec_size = shape_create()
    if pos == 0:
        continue


    if pos == 1:
        if count_1 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.rect((x_min, y_min), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min2, y_min2), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min3, y_min3), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name = "one" + new_name + str(count_1) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_1 = count_1 -1

    if pos == 2:
        if count_2 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.rect((x_min, y_min), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min2, y_min2), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min3, y_min3), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name =  "two" + new_name + str(count_2) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_2 = count_2 -1
        
    if pos == 3:
        if count_3 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.rect((x_min, y_min), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min2, y_min2), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.rect((x_min3, y_min3), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name =  "three" + new_name + str(count_3) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_3 = count_3 -1

```
