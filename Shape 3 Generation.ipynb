```python
import svgwrite
import random
from svglib.svglib import svg2rlg
from reportlab.graphics import renderPM
import math
TRIHEIGHT = math.sqrt(3)/2.0
```

```python
def draw_triangle(insert, size, fill, rotate=None):
    x, y = insert
    points = [insert, (x + size, y), ((x + size / 2.0), (y + size * TRIHEIGHT))]
    triangle = dwg.add(dwg.polygon(points, fill=fill, stroke=svgwrite.rgb(0, 0, 0, '%')))
    if rotate:
            triangle.rotate(rotate, center=insert)
```

```python
name = "shape.svg"
svg_size_width = 150
svg_size_height = 150
count_1 = 1000
rad_circ = 15
while(count_1 != 0):

    pos, x_min, y_min, x_max, y_max, x_min2, y_min2, rec_size = shape_create()
    if pos == 0:
        continue

    #rect circle
    if pos == 1:
        if count_1 == 0:
            continue
        dwg = svgwrite.Drawing(name, (svg_size_width, svg_size_height))
        dwg.add(dwg.rect((x_min2, y_min2), (rec_size,rec_size),
            stroke=svgwrite.rgb(0, 0, 0, '%'), 
            fill='white')
        )
        dwg.add(dwg.circle(center=(x_min + rad_circ,y_min + rad_circ),
            r=rad_circ, 
            stroke=svgwrite.rgb(0, 0, 0, '%'),
            fill='white')
        )
        dwg.save()
        drawing = svg2rlg(name)
        new_name = name.strip(".svg")
        new_name = "reccirc" + new_name + str(count_1) + ".png"
        renderPM.drawToFile(drawing, new_name, fmt="PNG")
        count_1 = count_1 -1


```
